```python
# Loops — recorrer estructuras
temperaturas = [32.5, 28.1, 30.7, 25.3, 29.8, 31.2, 27.4]

# Loop básico
for temp in temperaturas:
    print(f"Temperatura: {temp}°C")

print("---")

# Loop con condición
for temp in temperaturas:
    if temp > 30:
        print(f"{temp}°C — ALERTA: temperatura alta")
    else:
        print(f"{temp}°C — normal")

print("---")

# Enumerar con índice
for i, temp in enumerate(temperaturas):
    print(f"Día {i+1}: {temp}°C")
```

```python
def dividir(a, b):
    try:
        resultado = a / b
        return resultado
    except ZeroDivisionError:
        return "Error: no se puede dividir por cero"
    except TypeError:
        return "Error: los valores deben ser numeros"

print(dividir(10, 2))
print(dividir(10, 0))

print(dividir(10, "hola"))
```

```python
import random as rd
from datetime import datetime as dt

def generar_reporte_ventas(n_ventas):
    """
    Simula n ventas aleatorias y retorna un reporte.
    """
    ventas = [round(rd.uniform(50000, 2000000), 2) for _ in range(n_ventas)]
    
    try:
        promedio = sum(ventas) / len(ventas)
        mayor = max(ventas)
        menor = min(ventas)
        total = sum(ventas)
        ventas_altas = [v for v in ventas if v > 1000000]
        
        reporte = {
            "fecha": dt.now().strftime("%Y-%m-%d"),
            "total_ventas": n_ventas,
            "total_COP": round(total, 2),
            "promedio_COP": round(promedio, 2),
            "venta_mayor": mayor,
            "venta_menor": menor,
            "ventas_sobre_1M": len(ventas_altas)
        }
        return reporte
    
    except ZeroDivisionError:
        return "Error: debe haber al menos una venta"

reporte = generar_reporte_ventas(50)

for clave, valor in reporte.items():
    print(f"{clave}: {valor}")
```
