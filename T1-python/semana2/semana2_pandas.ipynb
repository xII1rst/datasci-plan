```python
import pandas as pd
import numpy as np

# Crear un DataFrame desde un diccionario
data = {
    "nombre":    ["Ana", "Carlos", "Diana", "Luis", "María"],
    "ciudad":    ["Bogotá", "Medellín", "Cali", "Valledupar", "Barranquilla"],
    "ventas":    [1200000, 850000, 2100000, 430000, 1750000],
    "mes":       ["Enero", "Enero", "Enero", "Enero", "Enero"]
}

df = pd.DataFrame(data)
print(df)
print(f"\nTipo: {type(df)}")
print(f"Shape: {df.shape}")
print(f"Columnas: {df.columns.tolist()}")
```

```python
# Exploración básica — los comandos que usarás siempre
print(df.head(3))       # primeras 3 filas
print("\n---")
print(df.tail(2))       # últimas 2 filas
print("\n---")
print(df.info())        # tipos de datos y valores nulos
print("\n---")
print(df.describe())    # estadísticas de columnas numéricas
```

```python
# Seleccionar una columna
print(df["ventas"])
print("\n---")

# Seleccionar varias columnas
print(df[["nombre", "ventas"]])
print("\n---")

# Filtrar filas por condición
print(df[df["ventas"] > 1000000])
print("\n---")

# Filtrar con múltiples condiciones
print(df[(df["ventas"] > 800000) & (df["ciudad"] != "Cali")])
```

```python
# Agregar columnas nuevas
df["ventas_millones"] = df["ventas"] / 1000000
df["cumple_meta"] = df["ventas"] > 1000000

# Modificar una columna existente
df["ciudad"] = df["ciudad"].str.upper()

# Aplicar una función a una columna
df["categoria"] = df["ventas"].apply(lambda x: "Alta" if x > 1500000 
                                     else "Media" if x > 800000 
                                     else "Baja")

print(df)
print(f"\nVendedores que cumplen meta: {df['cumple_meta'].sum()}")
```

```python
# Simular más datos para que groupby tenga sentido
data2 = {
    "vendedor": ["Ana", "Carlos", "Diana", "Ana", "Carlos", "Diana", "Ana"],
    "mes":      ["Ene", "Ene", "Ene", "Feb", "Feb", "Feb", "Mar"],
    "ventas":   [1200000, 850000, 2100000, 980000, 1100000, 1750000, 1500000]
}
df2 = pd.DataFrame(data2)

# Total de ventas por vendedor
print(df2.groupby("vendedor")["ventas"].sum())
print("\n---")

# Promedio de ventas por mes
print(df2.groupby("mes")["ventas"].mean().round(0))
print("\n---")

# Múltiples métricas a la vez
print(df2.groupby("vendedor")["ventas"].agg(["sum", "mean", "count"]))
```
